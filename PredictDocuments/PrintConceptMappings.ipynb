```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)

pandasDF = spark.sql("""select m.targetValue as concept, 
                    right(m.conceptMapUri,32) as context,
                    max(m.conceptmapversion) as version,
                    sourceValueSet,
                    sourceSystem,
                    SourceValue,
                    targetValueSet,
                    targetSystem,
                    targetValue
                    from ontologies_concepts.mappings m
                 where conceptMapUri like '%{CN}' 
                   
                 group by m.targetValue, m.conceptMapUri, sourceValueSet,
                    sourceSystem,
                    SourceValue,
                    targetValueSet,
                    targetSystem,
                    targetValue 
                 order by 1  """.format(**{
        'CN':contextNumber
    })).toPandas()  
```

```python
pandasDF.to_excel('conceptMappings.xlsx', encoding='utf-8')
```
